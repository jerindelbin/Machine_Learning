**DECISION TREE IMPLEMENTATION USING BREAST CANCER DATASET FROM SCKI-KIT LEARN**

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
cancer = pd.DataFrame(data = data.data, columns = data.feature_names)
cancer.head()
```

```python
cancer.shape
```

```python
cancer.columns
```

```python
#the target variable is not included in the dataset containing the features. 
#it is included in the dataset and needs to be addressed seperately from the df
data.target
```

```python
X = cancer
y = data.target
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'gini', max_depth = 7 )
```

```python
classifier.fit(X_train, y_train)
```

```python
y_prediction = classifier.predict(X_test)
y_prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_prediction, labels=[0,1])
```

```python
classifier.feature_importances_
```

```python
feature_importance = pd.DataFrame(classifier.feature_importances_, index = data.feature_names).sort_values(0, ascending=False)
feature_importance
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of given model is:",round(accuracy_score(y_test, y_prediction)*100,3),"%")
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure (figsize=(25,20))
_ = tree.plot_tree (classifier, feature_names = cancer.columns, class_names = {0 : 'Malignant', 1: 'Benign'}, filled = True, fontsize=12)
```
